# Working with histograms and random data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Uniform distribution

```python
xU=np.random.uniform(0,10,100000)
```

```python
fig, ax = plt.subplots(figsize = (5,5))
ax.hist(xU,bins=15,edgecolor='black',alpha=0.5)
plt.show()
```

## log-uniform distribution

```python
from loguniform import LogUniform
xLU=LogUniform(a=0.1,b=15).rvs(100000)
```

```python
fig, ax = plt.subplots(figsize = (5,5))
logbin=np.logspace(np.log10(xLU.min()),np.log10(xLU.max()),15)
ax.hist(xLU,bins=logbin,edgecolor='black')
plt.gca().set_xscale("log")
ax.set_xticks([0.2,0.5,1,2,10,20])
ax.set_xticklabels([0.2,0.5,1,2,10,20])
plt.show()
```

## Gaussian 'normal' distribution

```python
xG=np.random.normal(0,10,100000)
```

```python
fig, ax = plt.subplots(figsize = (5,5))
ax.hist(xG,bins=15,edgecolor='black',alpha=0.5)
plt.show()
```

## Cumulative distributions

### with `plt.hist`:

```python
fig, ax = plt.subplots(figsize = (5,5))
n,bins,pathes=ax.hist(xU,bins=15,edgecolor='black',alpha=0.5)
y=xU.cumsum()

plt.show()
```

```python
x=np.random.uniform(0,10,10000)
cumulative = np.cumsum(x)
```

```python
cumulative.shape
```

### With `numpy`:

```python
values, base = np.histogram(xU, bins=15)
cumulative = np.cumsum(values)
```

```python
fig, ax = plt.subplots(1,3,figsize = (15,5),sharex=True,sharey=True)
plt.subplots_adjust(wspace=0.02)

plt.subplot(ax[0])
plt.bar(base[:-1],100*values/values.sum(),color='red',alpha=0.5)
plt.ylabel('%total')
plt.title('uniform')
plt.subplot(ax[1])
plt.plot(base[:-1],100*cumulative/values.sum(),'b-')

plt.subplot(ax[2])
plt.bar(base[:-1],100*values/values.sum(),color='red',alpha=0.5)
plt.plot(base[:-1],100*cumulative/values.sum(),'b-')

plt.show()
```

```python
values, base = np.histogram(xLU, bins=15)
cumulative = np.cumsum(values)
```

```python
fig, ax = plt.subplots(1,3,figsize = (15,5),sharex=True)
plt.subplots_adjust(wspace=0.02)

plt.subplot(ax[0])
plt.bar(base[:-1],100*values/values.sum(),color='red',alpha=0.5)
plt.ylabel('%total')
plt.title('log-uniform')

plt.subplot(ax[1])
plt.plot(base[:-1],100*cumulative/values.sum(),'b-')

plt.subplot(ax[2])
plt.bar(base[:-1],100*values/values.sum(),color='red',alpha=0.5)
plt.plot(base[:-1],100*cumulative/values.sum(),'b-')



plt.show()
```

# Violin plots

```python
import numpy as np
import seaborn as sns
from loguniform import LogUniform


a=np.random.uniform(0,10,1000)
b=LogUniform(a=0.1,b=15).rvs(1000)
c=np.random.normal(5,1.5,1000)

t = np.concatenate((a,b,c))
t1=np.array(["A"]*len(a)+["B"]*len(b)+["C"]*len(c))
t_offset=np.concatenate((np.full(len(a),1.),np.full(len(b),2.),np.full(len(c),3.)))
data=np.array([t,t_offset])
color=['red','cadetblue','gold']
```

```python
fig,ax=plt.subplots(figsize=(5,3))
sns.violinplot(x=t,y=t1,palette=color,ax=ax,bw=.1, cut=0,inner='quartile',scale='count')
```

```python
t = np.concatenate((a,b))
t1=np.array(["A"]*len(a)+["B"]*len(b))
t_offset=np.concatenate((np.full(len(a),1.),np.full(len(b),2.)))
fig,ax=plt.subplots(1,2,figsize=(10,3))
test=sns.violinplot(x=t, y=t1,
              scale='count',
              hue=t_offset, split=True,
              inner='quartile',ax=ax[0])

test2 = sns.violinplot(x=t, y=np.full(len(t),"a"),
              scale='count',
              hue=t_offset, split=True,
              palette='seismic',
              inner='quartile')
```
