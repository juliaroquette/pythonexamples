# Multiple plots with `matplotlib`:

some references: https://matplotlib.org/tutorials/intermediate/gridspec.html


```python
import matplotlib.pyplot as plt
```

Create some random data:

```python
x=np.random.uniform(0,10,10000)
y=np.random.uniform(0,10,10000)
```

## with `plt.subplots`

Note that subplots only works well for plots with equal sizes but plotting very different sizes of plots with it is not a straightforward task.

It creates a grid of plots all at once such as they can be easily called inside a loop. 

### 4x1 grid

```python
fig, axes = plt.subplots(4,1, figsize=(8,8))
co=['r','k','b','g']
for i in range(4):
    ax=axes[i]
    plt.subplot(ax)
    plt.plot(x,y,'.',c=co[i])
```

### 4x2 grid

```python
fig, axes = plt.subplots(4,2, figsize=(8,8))
co=np.array([['r','k','b','g'],['y','gray','c','m']]).transpose()
for i in range(4):
    for j in range(2):
        ax=axes[i,j]
        plt.subplot(ax)
        plt.plot(x,y,'.',c=co[i,j])
```

### Multiple plots with a common continuous axis

```python
fig,axis= plt.subplots(1,4,figsize = (8,4))
co=np.array(['r','k','b','g'])


for i in range(4):
    ax=axis[i]
    plt.subplot(ax)
    plt.plot(x+i*10,y,'.',c=co[i])
    plt.setp(ax.get_yticklabels(), visible=False)

plt.show()
```

```python
fig,axis= plt.subplots(1,4,sharey=True,figsize = (8,4))
co=np.array(['r','k','b','g'])

plt.subplots_adjust(wspace=0., hspace=0.)

t=[0]
for i in range(4):
    ax=axis[i]
    plt.subplot(ax)
    plt.plot(x+i*10,y,'.',c=co[i])
    plt.setp(ax.get_yticklabels(), visible=False)
    plt.setp(ax.get_xticklabels(), visible=False) 
    plt.setp(ax.get_yticklabels(), visible=False)
    t.append((i+1)*10)
    plt.xlim(x.min()+i*10,x.max()+i*10)

fig.add_subplot(111, frameon=False,sharey=ax)

plt.xticks(t,t)
plt.xlabel("common X")
plt.ylabel("common Y")
plt.xlim(0,40)
plt.show()
```

## With `plt.subplot2grid`

As long as I understand it is more robust for creating a grid of plots with different sizes. But I it seems to create each plot per time, and so in case of repeated plots inside a loop, I need to repeat it.  **can this be improved?**

### 4x1 grid

```python
fig= plt.subplots(figsize = (8,8))
co=['r','k','b','g']
for i in range(4):
    ax=plt.subplot2grid((4,1),(i,0),colspan=1,rowspan=1)
    plt.plot(x,y,'.',c=co[i])
```

### 4x1 grid

```python
fig= plt.subplots(figsize = (8,8))
co=['r','k','b','g']
gs = gridspec.GridSpec(4, 1)
for i in range(4):
    ax =plt.subplot(gs[i])
    plt.plot(x,y,'.',c=co[i])
```

### 4x2 grid

```python
fig= plt.subplots(figsize = (8,8))
gs = gridspec.GridSpec(4, 2)
co=np.array([['r','k','b','g'],['y','gray','c','m']]).transpose()
for i in range(4):
    for j in range(2):
        ax = plt.subplot(gs[i,j])
        plt.plot(x,y,'.',c=co[i,j])
```

### various sizes

```python
fig= plt.subplots(figsize = (8,8))

gs = gridspec.GridSpec(4, 4)
co=np.array(['r','k','b','g','y','gray','c','m'])

axis=[]
axis.append(gs[0,:])
axis.append(gs[1,:-1])
axis.append(gs[1:,-1])
axis.append(gs[2,0])
axis.append(gs[2,1:3])


for i in range(5):
    ax = plt.subplot(axis[i])
    plt.plot(x,y,'.',c=co[i])
```

## Other useful subplot adjustments

### Space between plots:
To set the space between each plot in the grid (or remove it completely)

```python
fig= plt.subplots(figsize = (4,4))
co=np.array([['r','k','b','g'],['y','gray','c','m']]).transpose()
for i in range(4):
    for j in range(2):
        ax=plt.subplot2grid((4,4),(i,j),colspan=1,rowspan=1)
        plt.plot(x,y,'.',c=co[i,j])

#add:
plt.subplots_adjust(wspace=0., hspace=0.)
```

### Remove the tick labels:

```python
fig= plt.subplots(figsize = (4,4))
co=np.array([['r','k','b','g'],['y','gray','c','m']]).transpose()
for i in range(4):
    for j in range(2):
        ax=plt.subplot2grid((4,4),(i,j),colspan=1,rowspan=1)
        plt.plot(x,y,'.',c=co[i,j])
#add
        plt.setp(ax.get_xticklabels(), visible=False) 
        plt.setp(ax.get_yticklabels(), visible=False)
```

## Plot inside another plot:
 `add_axes[left, bottom, width, height]` can be used to add a plot inside another.
 
 All quantities are in fractions of figure width and height

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(x,y,'.',c='r')
plt.ylim(0,15) 
plt.xlim(0,15) 

ax2 = fig.add_axes([0.7, 0.7, 0.1, 0.1])
plt.plot(x,y,'.',c='b')
ax2.set_xlim(5,10)
ax2.set_ylim(5,10)


```

### Share axis:

To share the same y or x axis in all plots:

#### with `plt.subplots`

A single `plt.ylim` sets the range in the `y` axis for all plots

```python
fig, axes = plt.subplots(4,2, sharex=True, sharey=True, figsize=(4,4))
co=np.array([['r','k','b','g'],['y','gray','c','m']]).transpose()
for i in range(4):
    for j in range(2):
        ax=axes[i,j]
        plt.subplot(ax)
        plt.plot(x,y,'.',c=co[i,j])
plt.ylim(-3,15)        
```

#### with `plt.subplot2grid`

Only works if one of the axis if defined before:

```python
fig= plt.subplots(figsize = (4,4))

ax0=plt.subplot2grid((4,2),(0,0),colspan=1,rowspan=1)
co=np.array([['r','k','b','g'],['y','gray','c','m']]).transpose()
for i in range(4):
    for j in range(2):
        ax=plt.subplot2grid((4,4),(i,j),colspan=1,rowspan=1,sharex=ax0,sharey=ax0)
        plt.plot(x,y,'.',c=co[i,j])
plt.ylim(-5,15)        
```

## Miscellaneous

### Plot axis-tick-labels inside the plot:

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(x,y,'.',c='r')
ax.tick_params(axis="x",direction="in", pad=-15)
plt.ylim(-2,12) 
plt.xlim(-2,12) 
```

### Invert an axis

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(x,y,'.',c='r')
plt.gca().invert_yaxis() 
plt.ylim(-2,12) 
plt.xlim(-2,12) 
```

### labels on the left or top with a defined number of ticks

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(x,y,'.',c='r')
ax.yaxis.tick_right()
ax.xaxis.tick_top()
plt.locator_params(axis='y', nbins=2)
plt.ylim(-2,12) 
plt.xlim(-2,12) 
```

# logscale and manually define tick labels:

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(x,y,'.',c='r')


ax.set_yscale("log", nonposy='clip')
ax.set_yticks([1.1,2.5,7,10])
ax.set_yticklabels(['a','b','c','d'])

plt.ylim(-2,12) 
plt.xlim(-2,12) 

```

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(x,y,'.',c='r')


ax.set_yscale("log", nonposy='clip')
ax.set_yticks([0.001,0.1,1,5,10])
ax.set_yticklabels([0.001,0.1,1,5,10])

plt.ylim(-2,12) 
plt.xlim(-2,12) 

```

```python
fig, ax = plt.subplots(figsize=(4,4))


plt.plot(x,np.log10(y),'.',c='r')
ax.set_yscale("log", nonposy='clip')
ax.set_yticks([0.001,0.1,1,5,10])
ax.set_yticklabels([0.001,0.1,1,5,10])

plt.ylim(-2,12) 
plt.xlim(-2,12) 

```

# add_subplots

```python
fig = plt.figure(figsize=(10,15))

nplots=4
for i in range(4):
    
    ax = fig.add_subplot(2,nplots//3+1,(i+1))
    plt.plot(x,y,'.')
```

# hide top and right axis

```python
fig, ax = plt.subplots(figsize=(4,4))
plt.plot(x,y,'.',c='r')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.ylim(-2,12) 
plt.xlim(-2,12) 
```
